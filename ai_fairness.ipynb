```python
def distance(vector_1,vector_2):
    distance=0
    for i in range(len(vector_1)):
        #print(abs(vector_1[i]-vector_2[i]))
        distance+=abs(vector_1[i]-vector_2[i])
    return distance
```

```python
def update_centroids(datapoints,centroids,clusters):
    new_centroids=[i for i in range(len(centroids))]
    for i in range(len(clusters)):
        points=[]
        for point in clusters[i]:
            points.append((point,sum(datapoints[point])))
        points.sort(key=lambda x: x[1])
        median=len(clusters[i])//2
        new_centroid=points[median][0]
        new_centroids[i]=new_centroid
    return new_centroids
```

```python
iterations=10
for i in range(iterations):
    for k in range(5):
        clusters[k]=[]
    for index in range(len(datapoints)):
        point=datapoints[index]
        min_dist=200000000
        cluster=0
        for centroid in centroids:
            dist=distance(datapoints[centroid],datapoints[index])
            if min_dist>dist:
                min_dist=dist
                cluster=centroids.index(centroid)
        clusters[cluster].append(index)
    centroids=update_centroids(datapoints,centroids,clusters)
```
